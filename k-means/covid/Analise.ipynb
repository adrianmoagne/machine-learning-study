```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from biokit.viz import corrplot
import matplotlib.pyplot as plt
wine = "/home/adrian/Development/ml/Wine.csv"
```

```python
df = pd.read_csv(wine)


```

```python
matrix_corr = df.corr()
corr_graph = corrplot.Corrplot(matrix_corr)
corr_graph.plot(upper='ellipse', fontsize='x-large')
fig = plt.gcf()
fig.set_size_inches(13,8)
```

```python
atributos = df.columns
```

```python
for atributo in atributos:
    df[atributo] = (df[atributo]-min(df[atributo]))/(max(df[atributo])-min(df[atributo]))
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler= preprocessing.MinMaxScaler()
np_df = min_max_scaler.fit_transform(df)
```

```python
df = pd.DataFrame(np_df,columns=atributos)
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=4)
```

```python
cluster.fit(df)
```

```python
labels = cluster.labels_
print(labels)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = df['Color_Intensity'], y = df['Alcohol'], mode = 'markers', marker = dict(color=cluster.labels_.astype(float)),text=labels))
```
